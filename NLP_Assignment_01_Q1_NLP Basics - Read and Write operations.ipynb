```python
##TASK03## 
# USe the DOCX module to open the file and perform below tasks#

#pip install docx - in anaconda prompt
#or - pip install python-docx 

#import the docx module
import docx

#Python docx module allows users to manipulate docs by either manipulating the existing one or
#creating a new empty document and manipulating it.
#It is a powerful tool as it helps you to manipulate the document to a very large extend.

#read the word file
doc = docx.Document(r"C:\Users\YOGITA\Python\desktop\YS\NITW\Module-07 NLP\Assignment 01\NLP.docx")

#3.1# how many paragraphs are there in the file?

#print no. of paragraphs
paras = doc.paragraphs
print("Number of Paragraphs in the Doc:", len(paras))
# print(paras) #gives object (para) location
```

```python
#3.2#print the number of words in each paragraph

import re

para_num = 1

for para in paras:
    string = para.text
#     print(string)    ## prints the whole para as a string
    print("Number of words in para {}: {}".format(para_num, len(re.findall(r'\w+', string))))  #\w+ =all words
    para_num += 1
```

```python
##TASK04##

#read csv file
import pandas as pd
df = pd.read_csv(r"C:\Users\YOGITA\Python\desktop\YS\NITW\Module-07 NLP\Assignment 01\EmployeeDetails.csv")
print(df)
print("num of rows",len(df))  #no. of rows
```

```python
#split first and last names
first_names = []
last_names = []
for i in range(len(df)):
    split = df.iloc[i][0].split()  # iloc[i][0] - iloc[row_index][col_index] #col_index=0 for Name column
#     print(split)                                #split() - The split() method splits a string into a list.
    first_names.append(split[0])
    last_names.append(split[1])
df['First_Name'] = first_names
df['Last_Name'] = last_names
print(df)
```

```python

#delete Name column
print(df.columns)
df.drop(['Name '], axis=1, inplace=True)  #give column name in list!!!!
print(df)
```

```python
#increment salary by 10%
df['Salary'] = (df['Salary'] + df['Salary']*0.1)
print(df)
```

```python
#save to a new file
df.to_csv(r"C:\Users\YOGITA\Python\desktop\YS\NITW\Module-07 NLP\Assignment 01\Employee_Data.csv")

## overwrites the exisiting file, if file already exists.
```
